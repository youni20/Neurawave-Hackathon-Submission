# Weather API Example
Show how to fetch weather data (mock example)

```python
import requests

# Fetch weather data from SMHI (Swedish Meteorological and Hydrological Institute)
# Using SMHI Open Data API for weather forecast

# Example: Stockholm coordinates
lat = 59.3293
lon = 18.0686

url = f"https://opendata-download-metfcst.smhi.se/api/category/pmp3g/version/2/geotype/point/lon/{lon}/lat/{lat}/data.json"

response = requests.get(url)
weather_data = response.json()

print(f"Weather forecast for Stockholm (lat: {lat}, lon: {lon})")
print(f"Approved time: {weather_data['approvedTime']}")
print(f"Reference time: {weather_data['referenceTime']}")
print(f"\nFirst forecast:")
first_forecast = weather_data['timeSeries'][0]
print(f"Valid time: {first_forecast['validTime']}")
for param in first_forecast['parameters']:
    print(f"{param['name']}: {param['values'][0]} {param.get('unit', '')}")
```
